```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
col_names = ['I','II','III','aVR','aVL','aVF','V1','V2','V3','V4','V5','V6']
df1 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/file/Sample_SB.csv",names=col_names)
df2 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/file/Sample_SR.csv",names=col_names)
df3 = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/file/Sample_ST.csv",names=col_names)
srate = 500
timestamps = np.arange(0,10,1/srate)
x = timestamps
```

```python
fig, axs = plt.subplots(3,1,figsize=(10,8))
label_names = ['Test1','Test2','Test3']
axs[0].plot(x,df1[col_names[0]])
axs[1].plot(x,df2[col_names[0]])
axs[2].plot(x,df3[col_names[0]])

for i in range(3):
    axs[i].set_xlabel('Seconds')
    axs[i].set_xlabel('Microvolts')
    axs[i].grid(True)
    axs[i].set_title(label_names[i])
    plt.tight_layout()
```
